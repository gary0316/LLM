```python
import numpy as np
#匯入nimpy，用於處理數值運算和陣列操作
import pandas as pd
#匯入pandas庫，用於處理資料和分析
country=['Taiwan','Australia','Ireland','Australia','Ireland','Taiwan']
#建立country列表
age=[25,30,45,35,22,36]
#建立age列表
salary=[20000,32000,59000,60000,43000,52000]
#建立salary列表
dic={'Country':country,'Age':age,'Salary':salary}
#建立一個字典，包含三個鍵值與他們對應的值
data=pd.DataFrame(dic)
#使用pandas的DataFrame函數將dic轉換成DataFrame，並將結果賦值給data變數
data
#顯示data的內容
```

```python
from sklearn.preprocessing import LabelEncoder
#匯入scikit-learn庫中的LabelEncoder類別，用於對標籤進行編碼。
labelencoder = LabelEncoder()
#建立一個LabelEncoder的實例，用於將標籤轉換為數字。
data_le=pd.DataFrame(dic)
#建立一個包含國家、年齡和薪資資訊的 DataFrame
data_le['Country'] = labelencoder.fit_transform(data['Country'])
#使用LabelEncoder將DataFrame中的'Country'欄位的國家名稱進行編碼（轉換為數字），並將結果儲存在新的'Country'欄位中。
data_le
#顯示data_le的內容
```

```python
from sklearn.preprocessing import OneHotEncoder
#匯入scikit-learn庫中的OneHotEncoder類別，用於對標籤進行獨熱編碼。
from sklearn.compose import ColumnTransformer
#匯入scikit-learn庫中的ColumnTransformer類別，用於轉換DataFrame中的指定列。
ct = ColumnTransformer([("Country", OneHotEncoder(), [0])], remainder = 'passthrough')
'''
建立一個ColumnTransformer的實例
Country欄位（索引為0）應用OneHotEncoder。
remainder設為passthrough，這表示保留不在指定列上的所有其餘列。
'''
data_str_ohe=ct.fit_transform(data_le)
#使用ColumnTransformer將data_le中的Country欄位進行獨熱編碼，獨熱編碼的結果被儲存在data_str_ohe中。
pd.DataFrame(data_str_ohe)
#將獨熱編碼後的結果轉換成DataFrame並顯示。
```

```python
data_dum = pd.get_dummies(data)
#使用pd.get_dummies函式進行one-hot encoding，轉換成二元的dummy變數。
pd.DataFrame(data_dum)
#將轉換後的資料框轉換回DataFrame。
```
